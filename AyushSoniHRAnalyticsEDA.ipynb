```python

import pandas as pd
import numpy as np

city_day      = pd.read_csv('city_day.csv')
city_hour     = pd.read_csv('city_hour.csv')
station_day   = pd.read_csv('station_day.csv')
station_hour  = pd.read_csv('station_hour.csv')
stations      = pd.read_csv('stations.csv')

datasets = {'city_day' : city_day,'city_hour': city_hour,'station_day' : station_day,'station_hour': station_hour,'stations': stations}
for name, df in datasets.items():
    print(f"  {name} - {df.shape}")

```

```python
for name, df in datasets.items():
    print(f"\n{name}")
    display(df.describe(include='all').round(2))
```

```python



for name, df in datasets.items():
    missing = df.isnull().sum()
    missing_pct = (missing / len(df) * 100).round(2)
    missing_df = pd.DataFrame({
        'Count': missing,
        'Pct': missing_pct
    }).query('Count > 0').sort_values('Pct', ascending=False)
    print(f"\n{name}")
    display(missing_df)
```

```python
cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2',
       'O3', 'Benzene', 'Toluene', 'Xylene']

fig, axes = plt.subplots(4, 3, figsize=(18, 20))
axes = axes.flatten() # Flatten the 4x3 array of axes for easy iteration

for i, col in enumerate(cols):
    sns.histplot(city_day[col], kde=True, ax=axes[i]) # Add kde for density estimation
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2',
       'O3', 'Benzene', 'Toluene', 'Xylene']

fig, axes = plt.subplots(4, 3, figsize=(18, 20))
axes = axes.flatten() # Flatten the 4x3 array of axes for easy iteration

for i, col in enumerate(cols):
    sns.boxplot(y=city_day[col], ax=axes[i]) # Plot boxplot on current axis
    axes[i].set_title(f'Box Plot of {col}')
    axes[i].set_ylabel(col) # Set y-axis label

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2',
       'O3', 'Benzene', 'Toluene', 'Xylene']

print("Skewness for each pollutant in city_day:")
for col in cols:
    skewness_value = city_day[col].skew()
    print(f"{col}: {skewness_value:.2f}")
```

```python
for col in cols:
    city_day[col] = np.log1p(city_day[col])
```

```python
citycopy = city_day.drop(['City','Date','AQI_Bucket'], axis = 1)
```

```python
plt.figure(figsize=(12, 10)) # Set a good figure size for the heatmap
sns.heatmap(citycopy.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Pollutants and AQI')
plt.show()
```

PM2.5 and PM10 have strongest relationships with AQI which means that particulate matter is the primary driver of air pollution.

Nitrogen based pollutants like NO, NO2, NOx also show significant positive impact.

SO2 and NH3 also have moderate influence on AQI levels.

O3,  Benzene, Toluene, and Xylene have less impact on AQI.

AQI increases significantly when multiple pollutants increase simultaneously.

```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3',
       'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
print("zero values in each column")
for col in pollutants:
    print(f"{col} - {(mergeddf[col] == 0).sum()}")
```

```python
for col in pollutants:
    mergeddf[col] = mergeddf[col].replace(0, np.nan)
```

```python
for col in pollutants:
    print(f"{col} - Zero Values: {(mergeddf[col] == 0).sum()}")
```

Missing values

```python
for col in pollutants:
    print(f"{col} - {mergeddf[col].isna().mean()*100:.2f}%")
```
